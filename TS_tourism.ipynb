List down the top 10 districts that have the highest number of domestic visitors overall (2016 - 2019)?

```python
top_10_domestic_visitors=full_data.groupby(["district"]).agg({'domestic_visitors':sum}).reset_index()
top_10_domestic_visitors=top_10_domestic_visitors.sort_values(by='domestic_visitors',ascending=False)
top_10_domestic_visitors=top_10_domestic_visitors.reset_index().drop(columns=['index'])
top_10_domestic_visitors.head(10)
```

```python
visitors_by_year=full_data.groupby(["district","year"]).agg({'domestic_visitors':sum}).reset_index()
visitors_by_year=visitors_by_year.pivot(index='district',columns='year')
visitors_by_year.columns = visitors_by_year.columns.droplevel(level=0)

visitors_by_year=visitors_by_year.reset_index()

visitors_by_year.head(10)
```

# # 2.	List down the top 3 districts based on compounded annual growth rate (CAGR) of visitors between (2016 - 2019)?

```python
cagr=visitors_by_year
cagr["CAGR"]=round(((pow((cagr[2019]/cagr[2016]),(1 / 4)) - 1)*100),2)
cagr.sort_values(by="CAGR",ascending=False)
cagr=cagr.drop(index=10)
cagr=cagr.sort_values(by="CAGR",ascending=False)
top_3=cagr[["district","CAGR"]]
top_3.head(3)
```

## 3.	List down the bottom 3 districts based on compounded annual growth rate (CAGR) of visitors between (2016 - 2019)?

```python
cagr=cagr.sort_values(by="CAGR")

bottom_3=cagr[["district","CAGR"]]

bottom_3.head(3)
```

# 4.	What are the peak and low season months for Hyderabad based on the data from 2016 to 2019 for Hyderabad district?

```python
hyderabad=full_data[full_data['district']=='Hyderabad']
hyderabad.loc[:,'total_visitors'] = hyderabad["domestic_visitors"]+hyderabad["foreign_visitors"]
hyderabad=hyderabad.groupby('month').agg({'total_visitors':sum})
hyderabad.reset_index().sort_values(by='total_visitors',ascending=False)
```

## 5.	Show the top & bottom 3 districts with high domestic to foreign tourist ratio?

```python
district_wise = full_data
district_wise=district_wise.groupby('district').agg({'domestic_visitors':sum,'foreign_visitors':sum})

district_wise['ratio'] = (district_wise['domestic_visitors']//district_wise['foreign_visitors'])

top_3_ratio = district_wise.sort_values(by='ratio')
top_3_ratio.head(3)
```

## 8.	Estimate the projected revenue for Hyderabad in 2025 based on average spend per tourist 

```python
hyderabad = full_data[full_data["district"]=='Hyderabad']
hyderabad=hyderabad.groupby('year').agg({'domestic_visitors':sum,'foreign_visitors':sum}).reset_index()
hyderabad=hyderabad.transpose()
hyderabad.columns = hyderabad.iloc[0]
hyderabad=hyderabad.reset_index().drop(index=0)
hyderabad['CAGR'] = (hyderabad[2016]/ hyderabad[2019])**(1 / 3) - 1
hyderabad['projected_2025']=((1 + hyderabad['CAGR']) ** 6) * hyderabad[2019]
hyderabad
```
